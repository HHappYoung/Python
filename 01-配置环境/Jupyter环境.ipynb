# 二、Jupyter 快捷键

- Jupyter Notebook 有两种键盘输入模式。
  - 编辑模式，允许你往单元中键入代码或文本；这时的单元框线是绿色的。（左边是绿色）
  - 命令模式，键盘输入运行程序命令；这时的单元框线是灰色。（左边是蓝色）

- 命令模式(节选）
  - 基本控制:enter 切换到编辑模式，shift enter 运行当前单元格
  - 输入模式: m切换为markdown模式；y切换为代码块模式
  - 增添单元格:a在此单元格上面加一个，b为在此单元格下加一个。
  - 拷贝c 剪切x 粘贴v 删除d（连按两下） 撤销操作z 保存当前笔记本s
  - 内核操作: ii 中断 00重启
  - D,D : 删除选中的单元（连按两下D，可以撤销）
- 编辑模式（节选）
  - Ctrl+] : 缩进（可以整体缩进）
  - Ctrl+[ : 解除缩进  .
  - Ctrl+Enter : 运行本单元
  - Alt+Enter : 运行本单元，在下面插入一单元
  - Ctrl Shift+-- : 分割单元（根据鼠标光标位置来分）
  - Ctrl+/ :整体注释

### 1、基本语法
#### 1.1 多级标题

如果想将一段文字定义为标题，只需要在这段文字前面加上 `#`，**再在 `#` 后加一个空格**即可。还可增加二、三、四、五、六级标题，总共六级。每多一级 ，增加一个 `#` ，标题字号相应降低一级。

#### 1.2 列表

列表格式也很常用，它可以让你的文稿变得井井有条。在 Markdown 中，你只需要在文字前面加上 `-`或者`+`，然后保留一个空格就可以了；如果你希望是有序列表，在文字前面加上 `1. 或2. 或3.` 即可(后面需保留一个空格）。

## 设置 Jupyter 作图支持中文字体
+ 首先保障系统中存在中文字体，比如黑体
+ 如windows本机使用Jupyter,还需将黑体字文件（c:/windows/fonts/simhei.ttf）拷贝到  Lib\site-packages\matplotlib\mpl-data\fonts 文件夹中,或者添加字体文件夹的路径给python

````python
#rcParams本质上是字典
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] #正常显示中文（设置为黑体）
mpl.rcParams['font.serif'] = ['SimHei'] #正常显示中文
mpl.rcParams['axes.unicode_minus'] = False #正常显示负号，这个主要是因为黑体不能正常显示负号
````
